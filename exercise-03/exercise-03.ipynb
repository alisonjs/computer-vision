```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
roi_1 = cv.imread("images/roi_1.png")
roi_2 = cv.imread("images/roi_2.png")
roi_3 = cv.imread("images/roi_3.png")

mediana_altura = np.median([roi_1.shape[0], roi_2.shape[0], roi_3.shape[0]])
print(mediana_altura)
```

```python
mediana_comprimento = np.median([roi_1.shape[1], roi_2.shape[1], roi_3.shape[1]])
print(mediana_comprimento)
```

```python
face_1 = cv.resize(roi_1, (124, 124), interpolation=cv.INTER_LANCZOS4)
face_2 = cv.resize(roi_2, (124, 124), interpolation=cv.INTER_LANCZOS4)
face_3 = cv.resize(roi_3, (124, 124), interpolation=cv.INTER_LANCZOS4)
```

```python
face_1 = cv.cvtColor(face_1, cv.COLOR_BGR2RGB)
face_2 = cv.cvtColor(face_2, cv.COLOR_BGR2RGB)
face_3 = cv.cvtColor(face_3, cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(131)
plt.title("Pessoa 1")
plt.imshow(face_1)

plt.subplot(132)
plt.title("Pessoa 2")
plt.imshow(face_2)

plt.subplot(133)
plt.title("Pessoa 3")
plt.imshow(face_3)

plt.show()
```

```python
face_1_bgr = cv.cvtColor(face_1, cv.COLOR_RGB2BGR)
face_2_bgr = cv.cvtColor(face_2, cv.COLOR_RGB2BGR)
face_3_bgr = cv.cvtColor(face_3, cv.COLOR_RGB2BGR)

cv.imwrite("images/face_p_1.png", face_1_bgr)
cv.imwrite("images/face_p_2.png", face_2_bgr)
cv.imwrite("images/face_p_3.png", face_3_bgr)
```
