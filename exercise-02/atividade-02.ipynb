```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
classificador = cv.CascadeClassifier("/home/alison-ia/Documents/classificadores/haarcascade_frontalface_default.xml");
```

```python
woman_01 = cv.imread("woman-01.png")
woman_02 = cv.imread("woman-02.png")
woman_03 = cv.imread("woman-03.png")

woman_01 = cv.cvtColor(woman_01, cv.COLOR_BGR2RGB)
woman_02 = cv.cvtColor(woman_02, cv.COLOR_BGR2RGB)
woman_03 = cv.cvtColor(woman_03, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(woman_03)
```

```python
## Woman 01

woman_01_gray = cv.cvtColor(woman_01, cv.COLOR_RGB2GRAY)
woman_02_gray = cv.cvtColor(woman_02, cv.COLOR_RGB2GRAY)
woman_03_gray = cv.cvtColor(woman_03, cv.COLOR_RGB2GRAY)
```

```python
plt.imshow(woman_01_gray, cmap='gray')
```

```python
plt.imshow(woman_02_gray, cmap='gray')
```

```python
plt.imshow(woman_03_gray, cmap='gray')
```

```python
faces = []
```

```python
womans_gray = [woman_01_gray, woman_02_gray, woman_03_gray]
```

```python
for woman in womans_gray:
    faces.append(classificador.detectMultiScale(woman, 1.3, 5))
```

```python
faces
```

```python
images = [woman_01.copy(), woman_02.copy(), woman_03.copy()]
i = 0

for face in faces:
    for (x, y, w, h) in face:
        cv.rectangle(images[i], (x,y), (x+w,y+h), (255,255,0), 2)
        i+=1
```

```python
plt.imshow(images[2])
```

```python
for (x,y,w,h) in faces[0]:
    roi_1 = woman_01[y:y+h, x:x+w]

for (x,y,w,h) in faces[1]:
    roi_2 = woman_02[y:y+h, x:x+w]

for (x,y,w,h) in faces[2]:
    roi_3 = woman_03[y:y+h, x:x+w]
```
