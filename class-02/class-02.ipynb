```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("images/px-people.jpeg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
image_gray = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
```

```python
plt.imshow(image_gray, cmap='gray')
```

```python
classificador = cv.CascadeClassifier("classifiers/haarcascade_frontalface_default.xml");
```

```python
faces = classificador.detectMultiScale(image_gray, 1.3, 5)
```

```python
len(faces)
```

```python
faces[0]
```

```python
image_note = image.copy()
```

```python
for (x, y, w, h) in faces:
    cv.rectangle(image_note, (x,y), (x+w,y+h), (255,255,0), 2)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(image_note)
```

```python
face_image = 0

for (x, y, w, h) in faces:
    face_image += 1
    image_roi = image[y:y+h, x:x+w]
    image_roi = cv.cvtColor(image_roi, cv.COLOR_RGB2BGR)
    cv.imwrite("images/face_"+str(face_image)+".png", image_roi)
```
